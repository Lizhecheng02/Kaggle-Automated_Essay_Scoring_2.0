```python
from transformers import AutoTokenizer
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-v3-large")
```

```python
data = pd.read_csv("dataset/train.csv")
data.head()
```

```python
sentences = data["full_text"].tolist()
encoded_list = tokenizer.batch_encode_plus(sentences, padding=False)
token_counts = [len(tokens) for tokens in encoded_list["input_ids"]]
```

```python
print(min(token_counts), max(token_counts))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(
    token_counts,
    bins=range(min(token_counts), max(token_counts), 1),
    alpha=1.0,
    color="red"
)
plt.title("Distribution of Token Counts")
plt.xlabel("Number of Tokens")
plt.ylabel("Frequency")
plt.grid(axis="y", alpha=1.0)
plt.show()
```
